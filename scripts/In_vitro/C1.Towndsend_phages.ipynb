# Scrap the multi fasta files :

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from selenium.common.exceptions import TimeoutException
import os 
import warnings
warnings.filterwarnings("ignore") 

path_download = "/home/concha-eloko/Téléchargements"
path_out = "/media/concha-eloko/Linux/PPT_clean/in_vitro/Townsed/genomes"
path_driver = "/home/concha-eloko/.wdm/drivers/geckodriver/linux64/0.33/geckodriver"
path_destiny = "/media/concha-eloko/Linux/PPT_clean/in_vitro/Townsed"
service = Service(executable_path = path_driver)


def setup_driver():
    # Set up Selenium with a headless Firefox browser
    firefox_options = Options()
    firefox_options.add_argument("--headless")
    driver = webdriver.Firefox(service=service, options=firefox_options)
    return driver

def download_proteins(accessions):
    driver = setup_driver()
    try:
        for accession in accessions:
            url = f"https://www.ncbi.nlm.nih.gov/search/all/?term={accession}"
            driver.get(url)
            print("First URL:", driver.current_url)

            print(accession)
            # Click the 'Genome' button
            Genome_locator = (By.XPATH, '/html/body/main/div/div/div[10]/section/ul/li[5]/a/span[1]')
            Genome_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(Genome_locator))
            Genome_button.click()
            print("URL after the Genome button:", driver.current_url)
            print("Just clicked on button")

            # Click the 'Genomic sequences' button
            Genomic_sequences_button_locator = (By.XPATH, '/html/body/div[2]/main/div/div/div[1]/section[3]/div/div[1]/div[2]/span[2]/a')
            try:
                WebDriverWait(driver, 20).until(EC.presence_of_element_located(Genomic_sequences_button_locator))
                Genomic_sequences_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable(Genomic_sequences_button_locator))
                Genomic_sequences_button.click()
                print("URL after the Genomic sequences button clicked:", driver.current_url)
            except TimeoutException:
                print(f"Genomic sequences button not found for {accession}")
                print("Current URL:", driver.current_url)
                continue                      

            # Click the 'send to' button
            send_to_button_locator = (By.XPATH, '/html/body/div[1]/div[1]/form/div[1]/div[4]/div/div[1]/h4/a') 
            format_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(send_to_button_locator))
            format_button.click()
            print("Just clicked on the 'send to' button")

            # Click the 'Coding Sequences' button 
            coding_sequences_locator = (By.XPATH, '//*[@id="codeseq"]')
            coding_sequences_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(coding_sequences_locator))
            coding_sequences_option.click()
            print("Just clicked on the 'Coding Sequences' button")

            # Click the 'format' button
            #format_button_locator = (By.XPATH, '//*[@id="codeseq_format"]')
            #format_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(format_button_locator))
            #format_button.click()
            #fasta_option_locator = (By.XPATH, '//*[@id="fasta_cds_aa"]')
            #fasta_option = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(fasta_option_locator))
            #fasta_option.click()

            # Click the 'create file' button
            try:
                create_file_button_locator = (By.XPATH, '/html/body/div[1]/div[1]/form/div[1]/div[4]/div/div[1]/div[2]/div[2]/button')
                create_file_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(create_file_button_locator))
                create_file_button.click()
                print("Just clicked on the 'create file' button")
            except TimeoutException:
                #create_file_button_locator = (By.XPATH, '/html/body/div[1]/div[1]/form/div[1]/div[4]/div/div[1]/div[2]/div[2]/button')
                #create_file_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(create_file_button_locator))
                #create_file_button.click()
                print(f"Manually work with {accession}")
                #print("Just clicked on the alternative 'create file' button")
                continue
                
            # Move the downloaded file
            os.system(f"mv {path_download}/sequence.txt {path_out}/{accession}.multi.faa")
            print(f"Downloaded and moved {accession}.fasta")
            print("\n")

    except TimeoutException:
        print("Timeout occurred while waiting for elements.")
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()
```

```python
genomes = [accession for accession in open(f"{path_destiny}/genomes.towndsend.tsv").read().split("\n") if os.path.isfile(f"{path_out}/{accession}.multi.faa")== False]
download_proteins(genomes)
```
