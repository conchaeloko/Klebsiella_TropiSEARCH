```python
matrix.columns

strip_names = list(map(lambda x : "".join(x.split(" ")) ,list(matrix.columns)))
matrix.columns = strip_names
```

```python
hr_phages = {}

for phage in matrix.columns :
    phage_df = matrix[phage]
    kltypes = [kltype for kltype, score in phage_df.items() if score > 0 ]
    hr_phages[phage] = kltypes
```

```python
def clean_print(dico) :
    import pprint
    pp = pprint.PrettyPrinter(width = 150, sort_dicts = True, compact = True)
    out = pp.pprint(dico)
    return out 
    
clean_print(hr_phages)
```

```python
path_tax = "/media/concha-eloko/Linux/77_strains_phage_project/vConTACT2_77_phages_21042023"

taxonomy_info = pd.read_csv(f"{path_tax}/vcontact.results.f.csv", sep = "\t", header = 0).fillna("Outlier")
taxonomic_groups = {}

for _,row in taxonomy_info.iterrows() :
    if row["VC"] not in taxonomic_groups :
        tmp_list = []
        tmp_list.append(row["Phage_name"])
        taxonomic_groups[row["VC"]] = tmp_list
    else :
        taxonomic_groups[row["VC"]].append(row["Phage_name"])
```

```python
for cluster in taxonomic_groups :
    if len(taxonomic_groups[cluster])>1:
        print(cluster)
        print(taxonomic_groups[cluster])
        cluster_list = list(set(matrix.columns) & set(taxonomic_groups[cluster]))
        matrix_group= matrix[cluster_list]
        print(matrix_group.columns)
        plt.figure(figsize=(8,15))
        sns_plot= sns.heatmap(matrix_group,linecolor='white',
                              annot=None,cbar_ax=None, cbar =False, 
                              xticklabels=True,yticklabels=matrix_group.index, 
                              cmap='rocket_r',linewidths=1)
        plt.show()
        print("\n")
        
```

```python
import numpy as np

for cluster in taxonomic_groups :
    if len(taxonomic_groups[cluster])>1:
        hr_num = []
        print(cluster)
        print(taxonomic_groups[cluster])
        cluster_list = list(set(matrix.columns) & set(taxonomic_groups[cluster]))
        matrix_group= matrix[cluster_list]
        for phage in matrix_group.columns :
            hr = sum(matrix_group[phage])
            hr_num.append(hr)
        std = np.std(hr_num)
        print(std)
        print(hr_num)
        print("\n")
        #print(matrix_group)
```

# The Clustering 

```python
from sklearn.cluster import AgglomerativeClustering

matrix_t = matrix.T
matrix_t = matrix_t[matrix_t.sum(axis=1)>2]
matrix_t_t = matrix_t.T
```

```python
plt.figure(figsize=(16,16))
sns_plot= sns.heatmap(matrix_t_t,linecolor='red',annot=None,cbar_ax=None, cbar =False, xticklabels=True,yticklabels=matrix_t_t.index, cmap='bone')

```

```python
clustering = AgglomerativeClustering(n_clusters= None, distance_threshold=3.7, compute_distances=True,compute_full_tree=True).fit(matrix_t)
clustering.distances_
```
