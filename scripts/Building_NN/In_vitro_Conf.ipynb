```python
import os 
import json 
from collections import Counter
path_ppt = "/media/concha-eloko/Linux/PPT_clean"
```

```python
predictions_77 = json.load(open(f"{path_ppt}/77_phages.PPT_pred.json"))
```

```python
import pprint
pp = pprint.PrettyPrinter(width = 250, sort_dicts = True, compact = True)
pp.pprint(predictions_77)
```

```python
our_kl = [102 , 107 , 112 , 151 , 17, 2, 24 , 25 , 27 ,36, 64]
kltypes = [f"KL{n}" for n in our_kl]
```

```python
working_phages = []
for phage in predictions_77 : 
    for kltype in predictions_77[phage] : 
        if kltype in kltypes : 
            working_phages.append(phage)
            break
    else :
        pass
```

```python
counts = []
for phage in working_phages : 
    for kltype in predictions_77[phage] :
        if kltype in kltypes : 
            counts.append(kltype)

Counter(counts)
```

```python
working_phages_dico = {}
for phage in working_phages :
    working_phages_dico[phage] = predictions_77[phage]
pp.pprint(working_phages_dico)
```

```python
for phage in working_phages_dico : 
    print(phage , f"We have {len(working_phages_dico[phage])} predictions. Here are the relevant scores")
    for kltype in kltypes : 
        if kltype in working_phages_dico[phage] : 
            print(kltype , working_phages_dico[phage][kltype])
    print("\n")
```

```python
for kltype in kltypes :
    print(kltype)
    for phage in working_phages_dico :
        if kltype in working_phages_dico[phage] : 
            print(phage)
    print("\n")
```
