```python
import os 
import pandas as pd
from tqdm import tqdm 
import warnings
from collections import Counter
warnings.filterwarnings('ignore')

# local :
path_work = "/media/concha-eloko/Linux/PPT_clean"

DF_info = pd.read_csv(f"{path_work}/TropiGATv2.final_df.tsv", sep = "\t" ,  header = 0)
DF_lvl_0 = DF_info[~DF_info["KL_type_LCA"].str.contains("\\|")]
prophage_df = DF_lvl_0.drop_duplicates(subset = ["Phage"], keep = "first")
```

```python
dpos_df = DF_lvl_0.drop_duplicates(subset = ["index"], keep = "first")
dpos_df
```

```python
prophage_df
```

```python
dico_KLtypes = dict(Counter(prophage_df["KL_type_LCA"]))
sorted_KLtypes = sorted(dico_KLtypes, key=lambda x: int(x.split("KL")[1]))

clean_dico = {kltype : f"{dico_KLtypes[kltype]} ; {round(dico_KLtypes[kltype]*100/15046, 4)}%" for kltype in sorted_KLtypes}

```

```python

dico_KLtypes
```

```python
small_kltypes = []
for kl,eff in dico_KLtypes.items() :
    if eff < 21 :
         small_kltypes.append(kl)

print(len(small_kltypes), len(dico_KLtypes))
```

> Plot KL type distribution : 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensuring keys and values are in list format
keys = list(clean_dico.keys())
values = list(clean_dico.values())  # Directly getting values from the dictionary

# Choose a color for the bars
bar_color = '#c6aa58ff'  # You can specify any color you like

# Create the plot
plt.figure(figsize=(30, 10))
sns.barplot(x=keys, y=values, color=bar_color)

# Labeling
plt.xlabel('KL-locus')
plt.ylabel('Number of prophages')
#plt.title('Distribution of classes')
plt.xticks(rotation=90)  # Rotate x labels for better readability

# Save the plot
plt.savefig(f"/media/concha-eloko/Linux/papers/PPT/figures_draft/plot_prophage_counts.svg")

# Show the plot
plt.show()
```
