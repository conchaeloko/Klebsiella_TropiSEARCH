```python
import os 
import pandas as pd
from tqdm import tqdm
import json

#path_project = "/home/conchae/prediction_depolymerase_tropism/prophage_prediction/depolymerase_decipher/ficheros_28032023"
path_project = "/media/concha-eloko/Linux/PPT_clean/ficheros_28032023"
# ***************************************
# PPT :
dir_out_full = f"{path_project}/seekfold_PPT_full"
dir_out_dom = f"{path_project}/seekfold_PPT_domains"

ppt_missing_df = pd.read_csv(f"{path_project}/Missing_PPT.indices.tsv", names = ["indices"])

PPT_df = pd.read_csv("/media/concha-eloko/Linux/PPT_clean/PPT_results_DF.tsv", sep = "\t", header = 0 )
```

> Get the best domain : 

```python
# Fold infos :
#path_info = f"/home/conchae/depolymerase/depolymerase_fold.csv"
path_info = f"/media/concha-eloko/Linux/depolymerase_building/depolymerase_fold.csv"

info_df = pd.read_csv(path_info , sep = "\t", header = 0)

# Iterate the missing dpos :
header_seekfold = ["query","target","pident","alnlen","gapopen","qstart","qend","tstart","tend","bits","prob","aln_score"]
errors = {'A1l_00058_A_1_129_274.pdb','AF-A0A089GJA0-F1-model_v4.pdb','AF-A0A4V2M274-F1-model_v4.pdb',
          'AF-A0A7W9JCS6-F1-model_v4.pdb','AF-A0A658Y395-F1-model_v4.pdb'}

def find_match(results_df) :
    for _,row in results_df.iterrows() :
        if row["target"] not in errors :
            fold = info_df[info_df["ID"] == row["target"]]["Folds"].values[0]
            if fold not in ["jelly-roll", "Other"] :            
                return (row["query"] , row["alnlen"] , row["bits"] , fold)
                break
    else :
        return "Nada"

def get_domain_len(query) :
    if query == 0 :
        return 0
    else :
        start = int(query.split("A_")[1].split("_")[0])
        end = int(query.split("_")[-1].split(".pdb")[0])
        return end - start

def get_full_len_ppt(index) :
    seq = PPT_df[PPT_df["index"] == index]["sequence"].values[0]
    return len(seq)
    
def compute_score(metric, len_seq) :
    if len_seq == 0 :
        return 0
    else :
        score = (metric[1]**2 * metric[2]) / (len_seq)
        return score

def get_sequence(index) : 
    seq = PPT_df[PPT_df["index"] == index]["sequence"].values[0]
    return seq

def get_domain(index, query) :
    start = int(query.split("A_")[1].split("_")[0])
    end = int(query.split("_")[-1].split(".pdb")[0])
    seq = get_sequence(index)
    return seq[start:end]

ppt_dpo_domain = {}

for index_ppt in ppt_missing_df["indices"] :
    final_sequence = ""
    clean_index = index_ppt.split("__")[1]
    size_index = len(clean_index) + 2
    file_full = f"{dir_out_full}/{clean_index}.out"
    files_dom = [f"{dir_out_dom}/{file}" for file in os.listdir(dir_out_dom) if file[0:size_index] == f"{clean_index}_A"]
    # Get full results : 
    full_prot_df = pd.read_csv(f"{file_full}", sep = "\t" , names = header_seekfold)
    metric_full = find_match(full_prot_df)
    full_length = get_full_len_ppt(f"{index_ppt}")
    score_full = compute_score(metric_full, full_length)
    # Identify the best domain :
    best_domain = (0,0,0,0)
    for domain_out in files_dom :
        domain_df = pd.read_csv(f"{domain_out}", sep = "\t" , names = header_seekfold)
        if len(domain_df) > 0 :
            metric_dom = find_match(domain_df)
            if metric_dom != "Nada" :
                len_domain = get_domain_len(metric_dom[0])
                score_domain = compute_score(metric_dom, len_domain)
                score_best_domain = compute_score(best_domain, get_domain_len(best_domain[0]))
                if score_domain > score_best_domain :
                    best_domain = metric_dom
                else :
                    continue
    # Best domain VS full protein :
    len_domain = get_domain_len(best_domain[0])
    score_best_domain = compute_score(best_domain, len_domain)
    if score_full > score_best_domain :
        #print("Full won")
        #print(metric_full , full_length , score_full)
        #print(best_domain , len_domain , score_best_domain)
        final_sequence = get_sequence(index_ppt)
    else :
        #print("Domain won")
        #print(metric_full , full_length , score_full)
        #print(best_domain , len_domain , score_best_domain)
        final_sequence = get_domain(index_ppt , best_domain[0])
    #print("\n")
    # Get sequence : 
    a = {"full_seq" : get_sequence(index_ppt), "dpo_domain" : final_sequence}
    ppt_dpo_domain[index_ppt] = a
```

```python
with open(f"{path_project}/missing_dpos.ppt.tsv", "w") as outfile :
    for index in ppt_dpo_domain : 
        outfile.write(f"{index}\t{ppt_dpo_domain[index]['full_seq']}\t{ppt_dpo_domain[index]['dpo_domain']}\n")
        
```
