# 3. Affinity propagation
## a. PCA

```python
pca = PCA(.95)
pca_scaled_x = pca.fit_transform(embedding_df_comp)


    # Writing the figures :
features = range(pca.n_components_)
plt.figure(figsize=(16, 12))
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
#plt.savefig(f"{path_fig}/PCA.features.jpeg")
plt.show()
```

## b. Affinity Propagation 

```python
from sklearn.cluster import AffinityPropagation
from sklearn import metrics

af = AffinityPropagation(damping=0.95, preference=-0.1, random_state=123, max_iter=1000,verbose=True).fit(pca_scaled_x)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)
n_clusters_
```

## c. The clusters 

> Plotting the clusters

```python
import matplotlib.pyplot as plt
from itertools import cycle

plt.close("all")
plt.figure(figsize=(16, 12))
plt.clf()

colors = cycle("bgrcmykbgrcmykbgrcmykbgrcmyk")
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = pca_scaled_x[cluster_centers_indices[k]]
    plt.plot(pca_scaled_x[class_members, 0], pca_scaled_x[class_members, 1], col + ".")
    plt.plot(
        cluster_center[0],
        cluster_center[1],
        "o",
        markerfacecolor=col,
        markeredgecolor="k",
        markersize=14,
    )
    for x in pca_scaled_x[class_members]:
        plt.plot([cluster_center[0], x[0]], [cluster_center[1], x[1]], col)

plt.title("Estimated number of clusters: %d" % n_clusters_)
#plt.savefig(f"{path_fig}/AffPropagation.small.features.jpeg")
plt.show()
```

> Get information from inside the cluster 

```python
list_types = embedding_df["ID"].values.tolist()
cluster_dico={}
for i_label, cluster in enumerate(labels):
    if cluster not in cluster_dico :
        k_types=[]
        k_types.append(list_types[i_label])
        cluster_dico[cluster]=k_types
    else :
        cluster_dico[cluster].append(list_types[i_label])
    
```

```python
labels_results = {}
rows = []
for cluster in cluster_dico :
    k_types = list(map(lambda x : labels_df[labels_df["Protein"] == x]["Label"].values[0], cluster_dico[cluster]))
    ancestors = list(map(lambda x : labels_df[labels_df["Protein"] == x]["Ancestor"].values[0], cluster_dico[cluster]))
    sequences = list(map(lambda x : labels_df[labels_df["Protein"] == x]["Sequence"].values[0], cluster_dico[cluster]))
    pfams = list(map(lambda x : labels_df[labels_df["Protein"] == x]["PFAM"].values[0], cluster_dico[cluster]))
    proteins = cluster_dico[cluster]
    for (protein, label, ancestor, sequence, pfam) in zip (proteins,k_types,ancestors,sequences,pfams) :
        line = [cluster, protein, label, ancestor, sequence, pfam]
        rows.append(line)
AF_results = pd.DataFrame(rows, columns = ['Cluster', 'Protein','Label','Ancestor','Sequence','PFAM'])
```
