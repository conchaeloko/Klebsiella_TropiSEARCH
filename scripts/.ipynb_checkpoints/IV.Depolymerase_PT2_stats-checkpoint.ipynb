```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# generate some random data
X = np.random.randn(1000, 10)

# perform t-SNE analysis
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

# assign each point to a cluster
labels = np.random.randint(0, 5, 1000)

# create initial scatter plot
fig, ax = plt.subplots()
scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)

# define function to update plot with each frame
def update(frame):
    # change cluster assignments for some points
    labels[:100] = np.random.randint(0, 5, 100)
    # update colors of scatter plot
    scatter.set_array(labels)
    # return scatter plot object
    return scatter,

# create animation
ani = FuncAnimation(fig, update, frames=50, interval=100, blit=True)

# save animation to file
ani.save('clustering_animation.gif', writer='pillow')
```
